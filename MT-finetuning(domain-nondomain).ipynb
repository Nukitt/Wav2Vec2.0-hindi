# Tokenize the text

````python
from ilstokenizer import tokenizer
import codecs

def to_tokenize_and_lower(input_path, output_path):
  outfile = open(output_path, 'w')
  for line in codecs.open(input_path):
    line = line.strip()
    line = tokenizer.tokenize(line).lower()
    #print (line)
    outfile.write(line+'\n')
  outfile.close()
````

````python
to_tokenize_and_lower('/content/drive/MyDrive/data/SSMT/MT-test/train_1.src','/content/drive/MyDrive/data/SSMT/MT-test/train.src.tkn')
to_tokenize_and_lower('/content/drive/MyDrive/data/SSMT/MT-test/train_1.tgt','/content/drive/MyDrive/data/SSMT/MT-test/train.tgt.tkn')

to_tokenize_and_lower('/content/drive/MyDrive/data/SSMT/MT-test/valid_30.src','/content/drive/MyDrive/data/SSMT/MT-test/valid_1.src.tkn')
to_tokenize_and_lower('/content/drive/MyDrive/data/SSMT/MT-test/valid_30.tgt','/content/drive/MyDrive/data/SSMT/MT-test/valid_1.tgt.tkn')
````

````python
#plotting training loss
f=open('/content/drive/MyDrive/data/SSMT/MT-UD-D/train.log','r')
lines=f.readlines()
training_loss={}
for line in lines:
  if '| train |' in line:
    line=line.split('|')
    epoch=line[3].split(" ")[2]
    loss=line[4].split(" ")[2]
    training_loss[epoch]=loss

  
import matplotlib.pyplot as plt
keys = list(training_loss.keys())
values = [float(val) for val in list(training_loss.values())]

plt.plot(values)

# Add labels and title
plt.xlabel('epoch')
plt.ylabel('loss_value')
plt.title('Training loss variation')
````

````python
#calculalating bleu score

from nltk.translate.bleu_score import sentence_bleu
with open('/content/drive/MyDrive/data/SSMT/MT-test/w2v_final_TMT-D-D.txt','r') as predictions, open('/content/drive/MyDrive/data/SSMT/MT-test/valid_30.tgt','r') as actual:
  actuals=actual.readlines()
  predictions=predictions.readlines()
  count=1
  avg=0
  for actual, prediction in zip(actuals,predictions):
    actual=[actual[:-2].split()]
    prediction=prediction[:-2].split()
    bleu=sentence_bleu(actual, prediction,weights=(1, 0, 0, 0))
    avg=(avg+bleu)
    count+=1
    with open('/content/drive/MyDrive/data/SSMT/MT-test/w2v_final_TMT_DD_bleu.txt','a') as f:
      f.write('Actual : {}\n'.format(' '.join(actual[0])))
      f.write('Prediction : {} \n'.format(" ".join(prediction)))
      f.write("Bleu score: {}\n\n".format(bleu))
  with open('/content/drive/MyDrive/data/SSMT/MT-test/w2v_final_TMT_DD_bleu.txt','a') as f:
    f.write('Average_bleu : {}\n'.format(avg/count))
  print(avg/count)
````
